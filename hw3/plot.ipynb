```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

matplotlib.rcParams['figure.dpi'] = 300

def load_event_return(folder, key='Eval_AverageReturn'):
    ea = EventAccumulator(f'data/{folder}')
    ea.Reload()
    df = pd.DataFrame(ea.Scalars(key))
    return np.array(df['value'])

def plot_experiments(title, experiments, x_label='Iterations', y_label='Eval_AverageReturn'):
    plt.figure()
    for name in experiments:
        folder = experiments[name]
        value = load_event_return(folder)
        plt.plot(value)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True, axis='y', linestyle='--')
    plt.legend(experiments.keys(), loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
    plt.show()
    plt.close()
```

```python
experiments = {
    'exp1': 'q6a_sac_InvertedPendulum_100_InvertedPendulum-v4_16-10-2022_20-35-35',
    'exp2': 'q6a_sac_InvertedPendulum_100_InvertedPendulum-v4_16-10-2022_20-53-53'
}
plot_experiments('q6a_Inverted Pendulum', experiments)
```

```python
def load_event_return_another(folder, key='Train_BestReturn'):
    ea = EventAccumulator(f'data/{folder}')
    ea.Reload()
    df = pd.DataFrame(ea.Scalars(key))
    return np.array(df['value'])

def plot_experiments_another(title, experiments, x_label='Iterations', y_label='Train_BestReturn', key='Train_BestReturn'):
    plt.figure()
    for name in experiments:
        folder = experiments[name]
        value = load_event_return_another(folder,key)
        plt.plot(value)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True, axis='y', linestyle='--')
    plt.legend(experiments.keys(), loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
    plt.show()
    plt.close()
```
